```python
import requests
import pandas as pd
from pandas import json_normalize
```

```python
date="2016-01-01"
ticker="XOM"
#apikey=read("alpha access key.txt")

f=open("alpha access key.txt", "r");
apikey=f.read()
```

```python
# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&date='+date+'&outputsize=full&symbol='+ticker+'&apikey='+apikey
r = requests.get(url)
data = r.json()
```

```python
data
```

```python
df_data=pd.DataFrame(data['Time Series (Daily)'].items())
df_data
```

```python
df_data.rename(columns={0:'Date'},inplace=True)
df_data
```

```python
opn = [d.get('1. open') for d in df_data[1]]
df_data['Open']=opn
```

```python
high = [d.get('2. high') for d in df_data[1]]
df_data['High']=high
```

```python
low = [d.get('3. low') for d in df_data[1]]
df_data['Low']=low
```

```python
close = [d.get('4. close') for d in df_data[1]]
df_data['close']=close
```

```python
volume = [d.get('5. volume') for d in df_data[1]]
df_data['volume']=volume
```

```python
df_data=df_data.drop(columns=[1])
```

```python
df_data.to_csv('exxon.csv')
```
